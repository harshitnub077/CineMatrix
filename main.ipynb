```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# at last overview column is string now converting it into list so that
# i can concatenate it with the other columns

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# there is one issue in the keyword and cast that: MAJOR ISSUE
# 'Sam Worthington' to
# 'SamWorthington'

# sam and worthington are diff words as per the model it will confuse between
# there is also sam mendes
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
# or
# movies['cast'] = movies['cast'].apply(collapse)
# movies['crew'] = movies['crew'].apply(collapse)
# movies['genres'] = movies['genres'].apply(collapse)
# movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#here we are getting one issue that there are some same frequent words
#which are repeating like activity and activities , accident and accidently\
#we have to remove them all by method called stemming

# what stemming does
# ['loved','loving','love']
# ['love','love','love']

# p.stem('loved')
# p.stem('loving')
# p.stem('love') in all cases we will get love


import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)

```

```python
stem("in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron")
```

```python
new_df['tags']= new_df['tags'].apply(stem) #applied stemming to the tags column
```

```python
#we have total 4806 movies or 4806 vectors
#so we have to calculate the distance of the movies from one to another
#if distance big similarity less and distance less similarity big
```

```python
#euclidian distance = normal distance formula of points
# instead of that we will find the angle between the vectors(cosine distance)
#cosine distance = theeta inversely proportinal to distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors) #for each movie this will calculate the distance to all 4806 movies data
```

```python
# We got one problem there that pehli movie ka khudse similarity rate is 1 and so on tha
# but jab similarity wise sort karenege toh movie apna index loose kar dega
# jo bhi movie humne select ki hai useko pakadke rakhna padega
# using enumerate function
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6] #top 5 similar movies
# there is one problem which is solved
# the sorting is done on the basis of index ki last wali movie sabse upar aagyi
# using key = lambda x:x[1] hume 1st number ke hisab se sorting nahi karna hai 2nd wale se karna hai
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie.index]
  movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
  for i in movies_list:
    print(i[0])
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
